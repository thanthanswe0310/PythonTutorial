```python
#### Multiple functions can be applied at once.  

### DataFrameGroupBy.agg() allows you to pass a dictionary to your grouped DataFrame.
```

```python
# SELECT day, AVG(tip),COUNT(*)
# FROM tips
# GROUP BY day;
```

```python
tips.groupby("day").agg({"tip":"mean","day":"size"})
```

```python
### Grouping by more than one column is done by passing a list of columns to the groupby() method.


# SELECT smoker, day, COUNT(*), AVG(tip)
# FROM tips
# GROUP BY smoker, day;
```

```python
tips.groupby(["smoker","day"]).agg({"tip":["size","mean"]})
```

```python
### JOIN can be performed with join() or merge().
```

```python
df1 = pd.DataFrame({"key":["A","B","C","D"],"value":np.random.randn(4)})

df2 = pd.DataFrame({"key":["B","D","D","E"],"value":np.random.randn(4)})
```

```python
### inner join


### SELECT * FROM df1
### INNER JOIN df2
### ON df1.key =df2.key;
```

```python
### merge performs an inner JOIN by default

pd.merge(df1, df2, on="key")
```

```python
# LEFT outer join

# SELECT * FROM df1
# LEFT OUTER JOIN df2
# ON df1.key =df2.key;
```

```python
pd.merge(df1,df2,on="key",how="left")
```

```python
# RIGHT JOIN

# SELECT *
# FROM df1
# RIGHT OUTER JOIN df2
# ON df1.key =df2.key;
```

```python
pd.merge(df1, df2, on ="key",how="right")
```

```python
# FULL JOIN

# SELECT *
# FROM df1
# FULL OUTER JOIN df2
# ON df1.key= df2.key;
```

```python
pd.merge(df1, df2, on="key",how = "outer")
```

```python
# UNION

df1 = pd.DataFrame(

    {"city": ["Chicago","San Francisco","New York City"], "rank":range(1,4)}
)

df2 = pd.DataFrame({

    "city":["Chicago","Boston","Los Angeles"],"rank":[1,4,5]
})
```

```python
# SELECT city, rank
# FROM df1
# UNION ALL
# SELECT city, rank
# FROM df2;
```

```python
pd.concat([df1,df2])
```

```python
# SQL's UNION is similar to UNION ALL, however UNION will remove duplicate rows.

# SELECT city, rank
# FROM df1
# UNION
# SELECT city, rank
# FROM df2;
```

```python
pd.concat([df1,df2]).drop_duplicates()
```
