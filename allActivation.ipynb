```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
# Step Activation
```

```python
def step(x):
    if x >= 0:
        return 1
    else:
        return 0
    
```

```python
x = np.arange(-6,6,0.01)
step_output = [step(i) for i in x]

# Plotting 
fig,ax = plt.subplots(figsize=(9,5))
ax.plot(x,step_output, color ="#307EC7" ,linewidth =3, label ="step")

ax.legend(loc ="upper left", frameon =False)
fig.show()
```

```python
## Sigmoid Function

def sigmoid(x):
    s = (1/(1+np.exp(-x))) #Sigmoid function
    return s
```

```python
x = np.arange(-6,6,0.01)
#Plotting
fig,ax = plt.subplots(figsize =(9,5))
ax.plot(x,sigmoid(x),color ="#307EC7", linewidth =3, label="sigmoid")
ax.legend(loc="upper left", frameon='False')
fig.show()
```

```python
#Tanh Activation

def tanh(x):
    t = (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
    # t = np.tanh(x)
    return t
```

```python
x = np.arange(-6,6,0.01)
#Plotting

fig, ax = plt.subplots(figsize =(9,5))
ax.plot(x, tanh(x), color ="#307EC7",linewidth =3, label ="tanh")
ax.legend(loc="upper left", frameon=False)
fig.show()
```

```python
# Define ReLU function
def relu(x):
    return max(0, x)
```

```python
# Generate data
x = np.arange(-6, 6, 0.01)
relu_output = [relu(i) for i in x]

# Plotting
fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(x, relu_output, color="#307EC7", linewidth=3, label="ReLU")
ax.legend(loc="upper left", frameon=False)
plt.title("ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

```python
#Leaky ReLU


def leaky_relu(x):
    y =max(0.1*x,x)
    return y
```

```python
x = np.arange(-6,6,0.01)
leaky_relu_output = [leaky_relu(i) for i in x]
#Plotting 
fig, ax = plt.subplots(figsize=(9,5))
ax.plot(x, leaky_relu_output, color ="#307EC7", linewidth =3, label ="relu"
        )
ax.legend(loc ="upper left", frameon = False)
fig.show()
```

```python
# Softplus Activation

def softplus(x):
    s =np.log(1+np.exp(x))
    return s
```

```python
x = np.arange(-6,6,0.01)
#Plotting
fig,ax = plt.subplots(figsize=(9,5))
ax.plot(x,softplus(x), color ="#307EC7", linewidth = 3 , label="softplus")
ax.legend(loc ="upper left", frameon = False)
fig.show()
```

```python
# GeLU Activation

def sigmoid(x):
    s =(1/(1+np.exp(-x)))
    return s

def gelu(x):
    g = x* sigmoid(1.702*x)
    return g
```

```python
# Swish Activation Function

def swish(x):
    s = x* (1/(1+np.exp(-x)))
    return s
```

```python
x = np.arange(-6,6,0.01)
#Plotting
fig,ax = plt.subplots(figsize=(9,5))
ax.plot(x,swish(x), color ="#307EC7", linewidth = 3 , label="softplus")
ax.legend(loc ="upper left", frameon = False)
fig.show()
```

```python
# Mish Activation

def mish(x):
    s = x* (1/(1+np.exp(-x)))
    return s
```

```python
x = np.arange(-6,6,0.01)
#Plotting
fig,ax = plt.subplots(figsize=(9,5))
ax.plot(x,mish(x), color ="#307EC7", linewidth = 3 , label="softplus")
ax.legend(loc ="upper left", frameon = False)
fig.show()
```
